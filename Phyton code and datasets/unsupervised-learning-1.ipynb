```python
#! python
import numpy; 
import matplotlib.pyplot as plt 
import pandas as pd;
from sklearn.cluster import KMeans

#load data
profiles = pd.read_csv('user_profiles_normed.csv');

# create training dataset
training_data = profiles[["rounds","mean_game_score", "mean_total_score", "mean_challenge_level", "mean_OLM_use" , "mean_sfun", "mean_sdiff", "mean_sperf", "mean_srules"]]; 

# visualize training data
training_data.plot.scatter(x = "mean_challenge_level", y = "mean_total_score", s = 50);
plt.title("Scatter plot");
plt.show(); 

input_data = training_data[["mean_challenge_level", "mean_total_score"]]
model = KMeans(n_clusters=2, init= "k-means++", n_init = 10).fit(input_data);

colors = ["b", "r"]; 
plt.plot(model.cluster_centers_[0][0], model.cluster_centers_[0][1], 'o', color = 'black', markersize=11);  
plt.plot(model.cluster_centers_[1][0], model.cluster_centers_[1][1], 'o', color = 'black', markersize=11);  

for i, p in enumerate(numpy.asarray(input_data)):
    plt.plot(p[0], p[1], 'o', markersize=9, color = colors[model.labels_[i]]); 
plt.xlabel("challenge"); 
plt.ylabel("score"); 
plt.title("clustered data - inertia = " + str(round(model.inertia_,2))); 
plt.show(); 

# set width of bars
barWidth = 0.25
 
# set heights of bars
bars1 = model.cluster_centers_[0]; 
bars2 = model.cluster_centers_[1]; 
tlabels = ["challenge", "score"]

# Set position of bar on X axis
r1 = numpy.arange(len(bars1))
r2 = [x + barWidth for x in r1]

# Make the plot
plt.bar(r1, bars1, color='b', width=barWidth, edgecolor='white', label='cluster1')
plt.bar(r2, bars2, color='r', width=barWidth, edgecolor='white', label='cluster2')

# Add xticks on the middle of the group bars
plt.xlabel('cluster data (centroids)', fontweight='bold')
plt.xticks([r + 0.5*barWidth for r in range(len(bars1))], tlabels)
 
# Create legend & Show graphic
plt.legend()
plt.show()

# prediction of new data
new_data = [[0.37, 0.45], [0.29, 0.25], [0.27, 0.5], [0.35, 0.29], [0.23, 0.45],  [0.24, 0.53], [0.28, 0.42], [0.35, 0.34]]; 
new = model.predict(new_data); 
print("New data labels: " + str(new)); 

plt.plot(model.cluster_centers_[0][0], model.cluster_centers_[0][1], 'o', color = 'black', markersize=11);  
plt.plot(model.cluster_centers_[1][0], model.cluster_centers_[1][1], 'o', color = 'black', markersize=11);  
# print clustered training data
for i, p in enumerate(numpy.asarray(input_data)):
    plt.plot(p[0], p[1], 'o', markersize=9, color = colors[model.labels_[i]]); 

for i, n in enumerate(new): 
   plt.plot(new_data[i][0], new_data[i][1], 'x', color = colors[n], markersize = 9); 
   
plt.xlabel("challenge"); 
plt.ylabel("score"); 
plt.show(); 
```
